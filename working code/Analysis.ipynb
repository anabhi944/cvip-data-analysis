# Association Rule learning

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming the dataset contains transaction data with items listed in a single column
# Convert the data to the format required for market basket analysis
# This typically involves converting the transaction data into a one-hot encoded format
basket = (data.groupby(['BILL_ID', 'GRP'])['QTY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('BILL_ID'))


# Convert the quantities to binary values (1 or 0)
basket_sets = basket.applymap(lambda x: 1 if x > 0 else 0)

# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the association rules
print("Association Rules:")
print(rules)
```
